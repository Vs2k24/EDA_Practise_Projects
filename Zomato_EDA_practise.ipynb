```python
#now we see that country code column is present in both dataset zomato dataset and country code dataset
#we can combine both the df since country code is same in both
#we can combine using pandas function merge and on which feature are we combining the two=> Country Code
final_df=pd.merge(df,df_country,on='Country Code',how='left')
final_df.head(5)
```

```python
#now we will go inside the dataset and explore what we have
#to check datatypes
final_df.dtypes
```

```python
#we will apply the value counts method on the country col to check how many records are there for each country
final_df.Country.value_counts()
```

```python
#so we see from the above records are zomato is mainly(the main base) in india with 8652 records 
#so the max no of transactions are happening in india
```

```python
#if we just need the country name and not the counts use index method
country_names=final_df.Country.value_counts().index
country_names
```

```python
#if we just need the number of records for that country use values method
country_values=final_df.Country.value_counts().values
country_values
```

```python
#so now we have the index and values that means the country and records for each 
#we will plot a pie chart for top 3 countries which uses zomato based on the number of transcations 
plt.pie(country_values[:3],labels=country_names[:3],autopct='%1.3f%%')
#to display the pie chart use plt.show()
plt.show()
```

observation: Zomato maximum records or transactions are from india followed by us and then uk

```python
#lets see from which country the most rating is coming
#if we also include reset_index() we can convert this series into dataframe
final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index()
```

```python
#above you see that the last column is not named properly after you do reset_index so after doing reset_index() we also include the column name
#rename(columns={0:'Rating Count'})=> wherever the column name is 0 rename it with rating count
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
ratings
```

```python
'''
GROUP 1: Rating 4.5
  â””â”€â”€ Color: Dark Green
        â””â”€â”€ Text: Excellent
              â””â”€â”€ Pizza Hut

GROUP 2: Rating 4.0  
  â””â”€â”€ Color: Green
        â””â”€â”€ Text: Very Good
              â””â”€â”€ KFC, Subway

GROUP 3: Rating 2.5
  â””â”€â”€ Color: Orange
        â””â”€â”€ Text: Average
              â””â”€â”€ McDonald's, Burger King

GROUP 4: Rating 1.0
  â””â”€â”€ Color: Red
        â””â”€â”€ Text: Poor
              â””â”€â”€ Taco Bell
'''
```

so the observation we saw above is that more than 2000 people have not rated thats why aggregate count =0 then some have given poor rating some have given good rating etc. so what we can do here is that we can visualise this using a chart

## Observation
1. when rating is between 4.5 to 4.9-->excellent
2. when rating is between 4.0 to 4.4-->very good
3. when rating is between 3.5 to 3.9-->good
4. when rating is between 2.5 to 3.4-->average
5. when rating is between 1.8 to 2.4-->poor
6. when rating is 0-->not rated

```python
ratings.head()
```

```python
#to increase the fig size we will use matplotlib function rcParams
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x='Aggregate rating',y='Rating Count',data=ratings)
plt.show()
```

```python
#so in my xaxis i want aggreagate rating in y axis=>rating count(so we see how many people gave what rating) and we will also use the rating color
matplotlib.rcParams['figure.figsize']=(12,6)
#but if just write hue='rating color' the bins will be of different color it wont match the rating color thats why we need to map it
sns.barplot(x='Aggregate rating',y='Rating Count',hue='Rating color',data=ratings,palette=['blue','red','orange','yellow','green','green'])
plt.show()
```

## Observation
1. not rated is very high
2. maximum number of ratings is between 2.5 and 3.4

```python
#now we will do count plot we will get the frequency of each color
sns.countplot(x='Rating color',data=ratings,palette=['blue','red','orange','yellow','green','green'])
plt.show()
```

```python
'''
explaination of (final_df[final_df['Rating color']=='White'].groupby('Country').size().reset_index().raname(columns={0:'records'}))
1.'Rating color' == 'White': "Keep only rows where Rating color is White"
Think: "Show me only restaurants with White ratings"

2.groupby('Country'): "Group the filtered data by Country"
Think: "Now organize these White-rated restaurants by which country they're in"
Example grouping:
Group 1: All White-rated restaurants in India
Group 2: All White-rated restaurants in USA
Group 3: All White-rated restaurants in UK

3.size(): "Count how many restaurants in each group"
Think: "Count how many White-rated restaurants in India, how many in USA, etc."

4.reset_index(): "Make the Country column a regular column again"
Turns the grouped result into a clean dataframe

'''
```

Observations:
we see india has 2139 white rated restraunts => means that 2139 restraunts have got 0 rating. India is the country which has given 0 rating the most
